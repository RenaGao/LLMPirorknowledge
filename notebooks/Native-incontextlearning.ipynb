```python
#Urdu Dialogue

Scene: Two friends, Ayesha and Bilal, are discussing their plan to visit the library. They talk about the books they want to read and their study goals.

#Dialogue data 

Ayesha:
کیا تم نے نئی لائبریری دیکھی ہے؟
(Kya tum ne nayi library dekhi hai?)
“Have you seen the new library?”

Bilal:
ہاں، میں کل لائبریری گیا تھا۔
(Haan, main kal library gaya tha.)
“Yes, I went to the library yesterday.”

وہاں کا ماحول بہت اچھا ہے۔
(Wahan ka mahaul bohot acha hai.)
“The environment there is very good.”

Ayesha:
تم نے کون سی کتابیں لی؟
(Tum ne kaun si kitaabein li?)
“Which books did you take?”

Bilal:
میں نے تاریخ کی کچھ کتابیں لیں اور ایک نئی کہانی کی کتاب۔
(Main ne tareekh ki kuch kitaabein leen aur ek nayi kahani ki kitaab.)
“I took some history books and a new storybook.”

تم کون سی کتابیں پڑھنا چاہتی ہو؟
(Tum kaun si kitaabein parhna chahti ho?)
“Which books do you want to read?”

Ayesha:
میں اردو شاعری کی کتابیں پڑھنا چاہتی ہوں۔
(Main Urdu shaairi ki kitaabein parhna chahti hoon.)
“I want to read books on Urdu poetry.”

لیکن وقت کہاں ملتا ہے؟
(Lekin waqt kahaan milta hai?)
“But where is the time?”

Bilal:
اگر ہم وقت نکالیں تو سب ممکن ہے۔
(Agar hum waqt nikalain to sab mumkin hai.)
“If we make time, everything is possible.”

کیا تم کل میرے ساتھ لائبریری آنا چاہو گی؟
(Kya tum kal mere saath library aana chaho gi?)
“Would you like to come to the library with me tomorrow?”

Ayesha:
ہاں، یہ اچھا خیال ہے۔
(Haan, yeh acha khayaal hai.)
“Yes, that’s a good idea.”

کل میں تم سے دوپہر میں ملوں گی۔
(Kal main tum se dopehar mein miloon gi.)
“I’ll meet you tomorrow in the afternoon.”

Bilal:
ٹھیک ہے، پھر کل لائبریری میں ملاقات ہوگی۔
(Theek hai, phir kal library mein mulaqat hogi.)
“Alright, then we’ll meet at the library tomorrow.”

# Analysis of Grammatical Traits

	1.	Postpositions:
	•	“میں” (mein): Used as “in” in “لائبریری میں ملاقات ہوگی” (library mein mulaqat hogi).
	•	“کے” (ke): In “وہاں کا ماحول” (wahan ka mahaul), showing possession.
	2.	Gender Agreement:
	•	“دیکھی” (dekhi): Feminine verb form agreeing with “لائبریری” (library).
	•	“لی” (li): Feminine verb form agreeing with “کتابیں” (kitaabein).
	3.	Honorifics:
	•	“آپ” (aap) could be substituted for “تم” (tum) for a formal tone.
	•	The verbs adapt to match the respectful form when needed.
	4.	Oblique Case:
	•	“کتاب کی” (kitaab ki): Oblique case with the postposition “کی” (ki).
	5.	Adjective-Noun Agreement:
	•	“نئی لائبریری” (nayi library): Feminine adjective “نئی” (nayi) agreeing with “لائبریری” (library).
	•	“اردو شاعری کی کتابیں” (Urdu shaairi ki kitaabein): Adjectives and possessive case aligned with feminine “کتابیں” (kitaabein).
	6.	Compound Verbs:
	•	“وقت نکالیں” (waqt nikalain): Compound verb meaning “make time.”
	•	“دیکھا تھا” (dekha tha): Compound verb indicating a past action.
	7.	Flexible Word Order:
	•	“کل میں تم سے دوپہر میں ملوں گی” (Kal main tum se dopehar mein miloon gi): Flexible placement of “کل” (kal) at the beginning or end of the sentence.
	8.	Reduplication:
	•	While not explicitly in this dialogue, “سوچ سوچ کر” (soch soch kar) or “چاہتی چاہتی” (chahti chahti) could be added for emphasis.
```

```python
# Japanese dialogue 

Scene: Two friends, Sora and Aki, are discussing their plan to visit Kyoto. They talk about sightseeing, food, and transportation.

# Dialogue data 

Sora:
こんにちは、明日何をする予定ですか？
(Konnichiwa, ashita nani o suru yotei desu ka?)
“Hello, what are your plans for tomorrow?”

Aki:
明日は特に予定がありませんが、どうしてですか？
(Ashita wa toku ni yotei ga arimasen ga, doushite desu ka?)
“I don’t have any particular plans for tomorrow. Why do you ask?”

Sora:
京都に行きたいと思っています。一緒に行きませんか？
(Kyouto ni ikitai to omotteimasu. Issho ni ikimasen ka?)
“I’m thinking of going to Kyoto. Would you like to go together?”

Aki:
いいですね！京都は歴史的な場所が多いので、楽しそうです。
(Ii desu ne! Kyouto wa rekishiteki na basho ga ooi node, tanoshisou desu.)
“That sounds great! Kyoto has many historical places, so it seems fun.”

Sora:
それに、美味しい食べ物もたくさんありますよね。
(Sore ni, oishii tabemono mo takusan arimasu yo ne.)
“Also, there’s a lot of delicious food, right?”

Aki:
はい、京料理を食べるのが楽しみです。
(Hai, kyou-ryouri o taberu no ga tanoshimi desu.)
“Yes, I’m looking forward to eating Kyoto cuisine.”

Sora:
行き方を調べましたか？
(Ikikata o shirabemashita ka?)
“Have you checked how to get there?”

Aki:
はい、新幹線で行くのが一番早いと思います。
(Hai, shinkansen de iku no ga ichiban hayai to omoimasu.)
“Yes, taking the bullet train is the fastest, I think.”

Sora:
そうですね。でも、新幹線はちょっと高いかもしれません。
(Sou desu ne. Demo, shinkansen wa chotto takai kamoshiremasen.)
“That’s true. But the bullet train might be a little expensive.”

Aki:
それなら、高速バスを使ってもいいかもしれませんね。
(Sore nara, kousoku basu o tsukatte mo ii kamoshiremasen ne.)
“In that case, using the express bus might be a good idea.”

Sora:
そうしましょう！明日の朝、駅で待ち合わせましょうか？
(Sou shimashou! Ashita no asa, eki de machiawase mashou ka?)
“Let’s do that! Shall we meet at the station tomorrow morning?”

Aki:
はい、何時に集合しますか？
(Hai, nanji ni shuugou shimasu ka?)
“Sure, what time should we meet?”

Sora:
午前九時はどうですか？
(Gozen kuji wa dou desu ka?)
“How about 9 a.m.?”

Aki:
大丈夫です。じゃあ、また明日！
(Daijoubu desu. Jaa, mata ashita!)
“That works for me. See you tomorrow then!”

Sora:
はい、また明日！
(Hai, mata ashita!)
“Yes, see you tomorrow!”

# Analysis of Grammatical Traits

	1.	Particles:
	•	“は” (wa): Topic particle in “明日は” (ashita wa) indicating the topic is “tomorrow.”
	•	“を” (o): Object marker in “食べ物を” (tabemono o) indicating “food” as the object of eating.
	•	“に” (ni): Direction particle in “駅で待ち合わせましょうか” (eki de machiawase mashou ka) to indicate meeting at the station.
	2.	Politeness Levels:
	•	Polite speech with ”-ます” (masu) endings: “行きたいと思っています” (ikitai to omotteimasu) and “調べましたか” (shirabemashita ka).
	•	Casual forms could replace these for informal conversations.
	3.	Verb Conjugations:
	•	“行きたい” (ikitai): Expressing desire to go.
	•	“思っています” (omotteimasu): Continuous form indicating ongoing thought.
	•	“高いかもしれません” (takai kamoshiremasen): Expressing possibility with “かもしれません” (kamoshiremasen).
	4.	Topic Marking:
	•	“京都は歴史的な場所が多い” (Kyouto wa rekishiteki na basho ga ooi): Topic “Kyoto” (京都) is marked by “は” (wa).
	5.	Honorifics:
	•	Honorific expressions can be added depending on the social context. For example, “いらっしゃいますか” (irasshaimasu ka) instead of “行きますか” (ikimasu ka) for respectful tone.
	6.	Expressing Agreement:
	•	“そうしましょう” (Sou shimashou): Used to agree with a plan politely.
	7.	Conditional Clauses:
	•	“それなら、高速バスを使ってもいい” (Sore nara, kousoku basu o tsukatte mo ii): Conditional “それなら” (sore nara) is used to suggest an alternative.
	8.	Time Expressions:
	•	“午前九時” (Gozen kuji): Specific time is marked by “午前” (gozen) for “a.m.”
	9.	Politeness and Contextual Clues:
	•	Politeness levels match the neutral yet friendly tone of the conversation. This level is typical in conversations between equals who are not very casual.
```

```python
# korean dialogue 

Scene: Two friends, Minji and Jisoo, are discussing their plans for the weekend and what they might do together.

# Dialogue data 

Minji:
지수야, 이번 주말에 시간 있어?
(Jisoo-ya, ibeon jumal-e sigan isseo?)
“Jisoo, do you have time this weekend?”

Jisoo:
응, 있어. 왜?
(Eung, isseo. Wae?)
“Yes, I do. Why?”

Minji:
우리 같이 한강에 갈래? 요즘 날씨가 너무 좋아.
(Uri gachi Hangang-e gallae? Yojeum nalssi-ga neomu joa.)
“Shall we go to the Han River together? The weather is so nice these days.”

Jisoo:
좋아! 한강에서 뭐 할 거야?
(Joa! Hangang-eseo mwo hal geo-ya?)
“Sounds good! What are we going to do at the Han River?”

Minji:
치킨 먹으면서 자전거 탈까?
(Chikin meogeumyeonseo jajeongeo talkka?)
“Should we eat chicken while riding bikes?”

Jisoo:
완벽해! 자전거는 어디에서 빌릴 수 있어?
(Wanbyeokhae! Jajeongeoneun eodieseo billil su isseo?)
“Perfect! Where can we rent bikes?”

Minji:
한강공원 근처에 자전거 대여소가 있어.
(Hangang gongwon geuncheo-e jajeongeo daeyeoso-ga isseo.)
“There’s a bike rental shop near Hangang Park.”

Jisoo:
좋네. 그런데 몇 시에 만날까?
(Johne. Geureonde myeot sie mannalkka?)
“Sounds good. By the way, what time should we meet?”

Minji:
오전 열한 시는 어때?
(Ojeon yeolhan si-neun eottae?)
“How about 11 a.m.?”

Jisoo:
좋아. 그때 보자!
(Joa. Geuttae boja!)
“Great. See you then!”

Minji:
알겠어. 주말에 봐!
(Algesseo. Jumal-e bwa!)
“Got it. See you on the weekend!”

# Analysis of Grammatical Traits

	1.	Particles (조사):
	•	“에” (e): Indicates time or place in “이번 주말에” (ibeon jumal-e) (this weekend) and “한강에” (Hangang-e) (to the Han River).
	•	“에서” (eseo): Denotes the location of an action in “한강에서 뭐 할 거야?” (Hangang-eseo mwo hal geo-ya?) (What will we do at the Han River?).
	•	“는” (neun): Topic particle in “자전거는 어디에서” (jajeongeoneun eodieseo) (as for bikes, where?).
	2.	Politeness Levels (존댓말 vs. 반말):
	•	Casual speech (반말): Used between friends, e.g., “좋아!” (Joa!) (Good!) and “몇 시에 만날까?” (Myeot sie mannalkka?) (What time should we meet?).
	•	“알겠어” (Algesseo): Casual “Got it” contrasts with the polite form “알겠습니다” (Algesseumnida) used in formal situations.
	3.	Honorifics (높임말):
	•	Not explicitly used here because it’s a casual conversation. However, honorifics would replace casual speech for formal situations, e.g., “갈래요?” (Gallaeyo?) instead of “갈래?” (Gallae?).
	4.	Verb Conjugations:
	•	“갈래” (gallae): Suggesting a plan, “Shall we go?”
	•	“먹으면서” (meogeumyeonseo): Connecting actions, “while eating.”
	•	“빌릴 수 있어?” (billil su isseo): Ability form, “Can we rent?”
	5.	Topic Marking (주제 강조):
	•	“한강에서” (Hangang-eseo): The topic of the activity is introduced.
	•	“자전거는” (Jajeongeoneun): Topic particle “는” (neun) emphasizes bikes.
	6.	Conditional Suggestions (제안):
	•	“탈까?” (talkka): Suggesting an action with the conditional suffix “까” (kka), “Shall we ride?”
	7.	Time Expressions (시간 표현):
	•	“오전 열한 시” (ojeon yeolhan si): Morning 11 o’clock.
	8.	Expressions of Agreement and Excitement:
	•	“완벽해!” (Wanbyeokhae!): “Perfect!” emphasizes excitement.
	•	“좋아!” (Joa!): “Sounds good!” shows casual agreement.
```

```python
# Cantanese dialogue 

Scene: Two friends, Mei and Wing, are planning their weekend together and discussing activities.

#Dialogue data 

Mei:
喂，阿Wing，星期六有冇時間呀？
(Wai, a Wing, sing1 kei4 luk6 jau5 mou5 si4 gaan3 aa3?)
“Hey, Wing, do you have time on Saturday?”

Wing:
有呀，你想做啲咩呀？
(Jau5 aa3, nei5 soeng2 zou6 di1 me1 aa3?)
“Yes, what do you want to do?”

Mei:
我諗住去沙灘，最近天氣好正喎。
(Ngo5 lam2 zyu6 heoi3 saa1 taan1, zeoi3 gan6 tin1 hei3 hou2 zeng3 wo3.)
“I’m thinking of going to the beach. The weather has been great lately.”

Wing:
沙灘好呀！你諗住喺度玩咩呀？
(Saa1 taan1 hou2 aa3! Nei5 lam2 zyu6 hai2 dou6 waan2 me1 aa3?)
“The beach sounds good! What do you plan to do there?”

Mei:
游水同埋曬太陽囉，仲可以食啲嘢。
(Jau4 seoi2 tung4 maai4 saai3 taai3 joeng4 lo1, zung6 ho2 ji5 sik6 di1 je5.)
“Swim and sunbathe, and we can also eat something.”

Wing:
咁啱，我啱啱買咗個新泳衣！
(Gam3 ngaam1, ngo5 ngaam1 ngaam1 maai5 zo2 go3 san1 wing6 ji1!)
“Perfect, I just bought a new swimsuit!”

Mei:
咁好喎！星期六幾點見呀？
(Gam3 hou2 wo3! Sing1 kei4 luk6 gei2 dim2 gin3 aa3?)
“That’s great! What time should we meet on Saturday?”

Wing:
十點得唔得呀？太早會唔會好攰呀？
(Sap6 dim2 dak1 m4 dak1 aa3? Taai3 zou2 wui5 m4 wui5 hou2 gui6 aa3?)
“Is 10 a.m. okay? Would it be too early and tiring?”

Mei:
十點啱啱好啦，早啲出發會涼啲。
(Sap6 dim2 ngaam1 ngaam1 hou2 laa1, zou2 di1 ceot1 faat3 wui5 loeng4 di1.)
“10 a.m. is just right. Starting earlier will be cooler.”

Wing:
好呀，咁到時見啦！
(Hou2 aa3, gam3 dou3 si4 gin3 laa1!)
“Okay, see you then!”

Mei:
好，到時見！
(Hou2, dou3 si4 gin3!)
“Alright, see you then!”

# Analysis of Grammatical Traits

	1.	Particles (助詞):
	•	“呀” (aa3): Used at the end of a question or for softening, e.g., “有冇時間呀？” (jau5 mou5 si4 gaan3 aa3?) (Do you have time?).
	•	“囉” (lo1): Indicates a conclusion or suggestion, e.g., “曬太陽囉” (saai3 taai3 joeng4 lo1) (sunbathe then).
	•	“喎” (wo3): Expresses surprise or emphasis, e.g., “最近天氣好正喎” (zeoi3 gan6 tin1 hei3 hou2 zeng3 wo3) (The weather has been great lately).
	2.	Aspect Markers (動貌詞):
	•	“咗” (zo2): Marks a completed action, e.g., “我啱啱買咗個新泳衣” (ngo5 ngaam1 ngaam1 maai5 zo2 go3 san1 wing6 ji1) (I just bought a new swimsuit).
	3.	Topic-Comment Structure (主題評論):
	•	“沙灘好呀！” (Saa1 taan1 hou2 aa3!): “The beach sounds good!” sets the topic “沙灘” (saa1 taan1) and comments on it.
	•	“星期六幾點見呀？” (Sing1 kei4 luk6 gei2 dim2 gin3 aa3?): The topic “星期六” (sing1 kei4 luk6) introduces the time being discussed.
	4.	Sentence-Final Particles (語氣詞):
	•	“啦” (laa1): Suggests agreement or emphasis, e.g., “十點啱啱好啦” (sap6 dim2 ngaam1 ngaam1 hou2 laa1) (10 a.m. is just right).
	5.	Pronouns and Politeness:
	•	“你” (nei5): Informal “you,” appropriate for close friends.
	•	Polite equivalents like “您” (nei5) are used in formal or respectful contexts.
	6.	Time Expressions:
	•	“星期六” (sing1 kei4 luk6): Saturday.
	•	“十點” (sap6 dim2): 10 a.m.
	7.	Expressing Possibility or Suitability:
	•	“得唔得” (dak1 m4 dak1): Asking if something is okay, e.g., “十點得唔得呀？” (sap6 dim2 dak1 m4 dak1 aa3?) (Is 10 a.m. okay?).
	8.	Expressing Agreement:
	•	“好呀！” (Hou2 aa3!): Casual agreement, “Okay!”
	9.	Adverbs for Emphasis:
	•	“啱啱” (ngaam1 ngaam1): Emphasizes “just now” in “啱啱買咗” (ngaam1 ngaam1 maai5 zo2) (just bought).

```

```python
# Mandrain dialogue 

Scene: Two friends, Xiao Ming and Xiao Li, are discussing their weekend plans and what they can do together at the park.

Dialogue

Xiao Ming:
小李，这个周末你有空吗？
(Xiǎo Lǐ, zhè gè zhōu mò nǐ yǒu kòng ma?)
“Xiao Li, are you free this weekend?”

Xiao Li:
有啊！你有什么计划吗？
(Yǒu a! Nǐ yǒu shén me jì huà ma?)
“Yes, I do! Do you have any plans?”

Xiao Ming:
我想去公园玩儿，最近天气很好。
(Wǒ xiǎng qù gōng yuán wánr, zuì jìn tiān qì hěn hǎo.)
“I want to go to the park; the weather has been great recently.”

Xiao Li:
好主意！你想做什么？
(Hǎo zhǔ yì! Nǐ xiǎng zuò shén me?)
“Good idea! What do you want to do?”

Xiao Ming:
可以骑自行车，也可以野餐。
(Kě yǐ qí zì xíng chē, yě kě yǐ yě cān.)
“We can ride bikes or have a picnic.”

Xiao Li:
听起来不错！自行车在哪儿租啊？
(Tīng qǐ lái bú cuò! Zì xíng chē zài nǎr zū a?)
“That sounds good! Where can we rent bikes?”

Xiao Ming:
公园门口有一个租自行车的地方，我上次去过。
(Gōng yuán mén kǒu yǒu yī gè zū zì xíng chē de dì fāng, wǒ shàng cì qù guò.)
“There’s a place to rent bikes at the park entrance; I went there last time.”

Xiao Li:
太好了！我们几点见？
(Tài hǎo le! Wǒ men jǐ diǎn jiàn?)
“Great! What time should we meet?”

Xiao Ming:
上午十点怎么样？
(Shàng wǔ shí diǎn zěn me yàng?)
“How about 10 a.m.?”

Xiao Li:
没问题！到时候见！
(Méi wèn tí! Dào shí hòu jiàn!)
“No problem! See you then!”

Xiao Ming:
好，周末见！
(Hǎo, zhōu mò jiàn!)
“Alright, see you this weekend!”

Analysis of Grammatical Traits

	1.	Particles (语气助词):
	•	“吗” (ma): Used for yes/no questions, e.g., “你有空吗？” (Nǐ yǒu kòng ma?) (Are you free?).
	•	“啊” (a): Adds emphasis or softness, e.g., “在哪儿租啊？” (Zài nǎr zū a?) (Where can we rent bikes?).
	•	“了” (le): Indicates a change of state or completion, e.g., “太好了！” (Tài hǎo le!) (That’s great!).
	2.	Aspect Markers (动态助词):
	•	“过” (guò): Indicates past experience, e.g., “我上次去过” (Wǒ shàng cì qù guò) (I went there last time).
	3.	Topic-Comment Structure (话题-评论):
	•	“自行车在哪儿租啊？” (Zì xíng chē zài nǎr zū a?): “Bikes” is the topic, and “where to rent them” is the comment.
	4.	Verb Complements (结果补语):
	•	“听起来” (tīng qǐ lái): “Sounds like,” a complement indicating how something seems.
	5.	Politeness Levels:
	•	Casual but polite tone appropriate for close friends, e.g., “你有空吗？” (Nǐ yǒu kòng ma?) uses polite language.
	6.	Conjunctions (连接词):
	•	“也” (yě): Indicates “also,” e.g., “也可以野餐” (Yě kě yǐ yě cān) (We can also have a picnic).
	•	“或者” (huò zhě): Could replace “也” to express alternatives in formal contexts.
	7.	Time Expressions:
	•	“上午十点” (shàng wǔ shí diǎn): 10 a.m., a specific time.
	•	“最近” (zuì jìn): Recently.
	8.	Expressions of Agreement and Excitement:
	•	“太好了！” (Tài hǎo le!): Expresses enthusiasm.
	•	“没问题！” (Méi wèn tí!): “No problem!” shows agreement.
```

```python
# Thai dialogue 

Scene: Two friends, Nuch and Lek, are planning their weekend together and discussing activities.

# Dialogue data 

Nuch:
เล็ก วันเสาร์นี้ว่างไหม?
(Lék wan sǎo níi wâang mái?)
“Lek, are you free this Saturday?”

Lek:
ว่างสิ มีอะไรเหรอ?
(Wâang sì. Mii à-rai rǒe?)
“I’m free. What’s up?”

Nuch:
เราไปสวนจตุจักรกันดีไหม? ฉันอยากไปเดินเล่น ซื้อของหน่อย
(Rao bpai sǔan Jàtùjàk gan dii mái? Chǎn yàak bpai doen lên, sǔue khǎawng nòi.)
“Shall we go to Chatuchak Park? I’d like to stroll around and shop a little.”

Lek:
ดีมากเลย! ช่วงนี้อากาศดีด้วย
(Dii mâak loei! Chûang níi aa-gàat dii dûuai.)
“That’s great! The weather is nice these days too.”

Nuch:
ใช่ เราไปเช้าๆ จะได้ไม่ร้อน
(Châi, rao bpai cháo cháo jà dâi mâi rón.)
“Yes, let’s go early so it won’t be too hot.”

Lek:
แล้วจะนัดเจอกันกี่โมงดี?
(Láew jà nát jer gan gìi moong dii?)
“What time should we meet?”

Nuch:
แปดโมงครึ่งได้ไหม?
(Bpàet moong khrûeng dâi mái?)
“Is 8:30 okay?”

Lek:
ได้สิ ฉันจะเตรียมของไปกินด้วยนะ
(Dâi sì. Chǎn jà dtriiam khǎawng bpai gin dûuai ná.)
“Sure. I’ll prepare some food to bring too.”

Nuch:
ดีจัง! ฉันจะเตรียมเครื่องดื่ม
(Dii jang! Chǎn jà dtriiam khrûeang dûuem.)
“That’s great! I’ll prepare some drinks.”

Lek:
ตกลง เจอกันวันเสาร์เช้านะ
(Dtòk-long, jer gan wan sǎo cháo ná.)
“Alright, see you Saturday morning.”

Nuch:
โอเค เจอกัน!
(O-kay, jer gan!)
“Okay, see you!”

# Analysis of Grammatical Traits

	1.	Particles (คำลงท้าย):
	•	“ไหม” (mái): Used to form yes/no questions, e.g., “ว่างไหม?” (wâang mái) (Are you free?).
	•	“สิ” (sì): Adds emphasis or certainty, e.g., “ว่างสิ” (wâang sì) (I’m free for sure).
	•	“นะ” (ná): Softens statements, e.g., “เจอกันนะ” (jer gan ná) (See you).
	2.	Aspect Markers (คำแสดงเวลา):
	•	“จะ” (jà): Marks future tense, e.g., “ฉันจะเตรียม” (chǎn jà dtriiam) (I will prepare).
	•	“ได้” (dâi): Indicates possibility, e.g., “ได้ไหม?” (dâi mái) (Is it okay?).
	3.	Topic-Comment Structure (หัวเรื่อง-ความคิดเห็น):
	•	“วันเสาร์นี้ว่างไหม?” (wan sǎo níi wâang mái?): The topic “วันเสาร์นี้” (wan sǎo níi) introduces the time, and the comment “ว่างไหม?” (wâang mái) asks if free.
	4.	Politeness Levels:
	•	Casual tone is used with “เล็ก” (Lék) and “ฉัน” (chǎn), suitable for close friends.
	•	Polite particles like “ครับ” (khráp) or “ค่ะ” (khâ) would be added in formal settings.
	5.	Verb Serialization:
	•	“ไปเดินเล่น ซื้อของหน่อย” (bpai doen lên, sǔue khǎawng nòi): Verbs “go,” “stroll,” and “buy” are serialized to describe multiple actions.
	6.	Time Expressions:
	•	“แปดโมงครึ่ง” (bpàet moong khrûeng): 8:30 a.m.
	•	“ช่วงนี้” (chûang níi): Recently or this period.
	7.	Expressions of Agreement and Excitement:
	•	“ดีมากเลย!” (Dii mâak loei!): Expresses excitement.
	•	“ดีจัง!” (Dii jang!): Shows positive feeling.
	8.	Conditional Suggestions:
	•	“จะได้ไม่ร้อน” (jà dâi mâi rón): “So it won’t be hot,” showing purpose.
	9.	Conjunctions (คำสันธาน):
	•	“แล้ว” (láew): Connects ideas, e.g., “แล้วจะนัดเจอกันกี่โมงดี?” (Láew jà nát jer gan gìi moong dii?) (What time should we meet?).
```

```python
# Malay dialogue 

Scene: Two friends, Aiman and Farah, are planning to visit the mall over the weekend and discuss what to do there.

# Dialogue data 

Aiman:
Farah, awak ada rancangan hujung minggu ni?
(Farah, awak ada rancangan hujung minggu ni?)
“Farah, do you have any plans this weekend?”

Farah:
Tak ada apa-apa pun. Kenapa?
(Tak ada apa-apa pun. Kenapa?)
“No, nothing at all. Why?”

Aiman:
Saya nak pergi ke pusat membeli-belah baru di bandar. Awak nak ikut?
(Saya nak pergi ke pusat membeli-belah baru di bandar. Awak nak ikut?)
“I want to go to the new shopping mall in the city. Do you want to come along?”

Farah:
Boleh juga! Ada apa yang menarik di sana?
(Boleh juga! Ada apa yang menarik di sana?)
“Sure! What’s interesting there?”

Aiman:
Khabarnya ada banyak kedai baru, dan saya dengar food court dia besar.
(Khabarnya ada banyak kedai baru, dan saya dengar food court dia besar.)
“I heard there are many new shops, and their food court is big.”

Farah:
Bagus tu! Kita boleh makan tengah hari di sana. Lepas tu, nak tengok wayang?
(Bagus tu! Kita boleh makan tengah hari di sana. Lepas tu, nak tengok wayang?)
“That’s great! We can have lunch there. After that, shall we watch a movie?”

Aiman:
Setuju! Wayang apa yang awak nak tengok?
(Setuju! Wayang apa yang awak nak tengok?)
“Agreed! What movie do you want to watch?”

Farah:
Saya rasa filem baru tu, yang semua orang cakap pasal.
(Saya rasa filem baru tu, yang semua orang cakap pasal.)
“I think the new movie that everyone’s talking about.”

Aiman:
Okey, kita tengok jadual wayang nanti. Jumpa pukul 10 pagi?
(Okey, kita tengok jadual wayang nanti. Jumpa pukul 10 pagi?)
“Okay, let’s check the movie schedule later. Meet at 10 a.m.?”

Farah:
Boleh! Jangan lupa belikan tiket dulu.
(Boleh! Jangan lupa belikan tiket dulu.)
“Sure! Don’t forget to buy the tickets first.”

Aiman:
Baik, saya uruskan. Jumpa nanti!
(Baik, saya uruskan. Jumpa nanti!)
“Alright, I’ll take care of it. See you later!”

Farah:
Okey, jumpa nanti!
(Okey, jumpa nanti!)
“Okay, see you later!”

# Analysis of Grammatical Traits

	1.	Particles (Partikel):
	•	“pun”: Used for emphasis, e.g., “Tak ada apa-apa pun.” (Nothing at all).
	•	“ke”: Indicates direction, e.g., “pergi ke pusat membeli-belah” (go to the mall).
	2.	Aspect Markers (Penanda Aspek):
	•	“nak”: Informal future marker, e.g., “Saya nak pergi” (I want to go).
	•	“dengar”: Implied past aspect in “saya dengar food court dia besar” (I heard their food court is big).
	3.	Topic-Comment Structure (Struktur Topik-Komen):
	•	“Wayang apa yang awak nak tengok?” (What movie do you want to watch?): Topic “Wayang apa” introduces the subject, and “awak nak tengok” comments on it.
	4.	Politeness Levels:
	•	Formal tone with “saya” (I) and “awak” (you) is polite but casual, suitable for friendly conversations.
	•	Politeness can be enhanced with “Encik” or “Cik” for formal contexts.
	5.	Verb Serialization:
	•	“Makan tengah hari di sana. Lepas tu, nak tengok wayang?” (Have lunch there. After that, shall we watch a movie?): Actions are listed sequentially.
	6.	Conjunctions (Kata Hubung):
	•	“dan”: Connects clauses, e.g., “banyak kedai baru, dan saya dengar” (many new shops, and I heard).
	•	“Lepas tu”: Informal for “after that.”
	7.	Time Expressions:
	•	“hujung minggu ni” (this weekend).
	•	“pukul 10 pagi” (10 a.m.).
	8.	Expressions of Agreement:
	•	“Setuju!” (Agreed!).
	•	“Boleh!” (Sure!).
	9.	Conditional Suggestions:
	•	“Kita tengok jadual wayang nanti.” (Let’s check the movie schedule later): Indicates a planned action.
	10.	Adjectives for Excitement:
	•	“Bagus tu!” (That’s great!) expresses enthusiasm.
```
